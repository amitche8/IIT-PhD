```python
import openstudio
```

```python

model = openstudio.model.Model()

space_1 = openstudio.model.Space(model)
space_1.setName('Happy Zone')

space_2 = openstudio.model.Space(model)
space_2.setName('UNIT: Sad Zone')

# Create Fridge Schedule
fridge_schedule = openstudio.model.ScheduleCompact(model)
fridge_schedule.setName('Fridge Schedule')
fridge_schedule.setToConstantValue(1.0)

# Add IHGs to Spaces
for space in model.getSpaces():
    name = space.name()
    area = space.floorArea()

    print(str(name)+'__'+str(area))
    if 'UNIT' in str(name):

        # Create fridge definition
        fridge = openstudio.model.ElectricEquipmentDefinition(model)
        fridge.setName('Fridge')
        fridge.setDesignLevelCalculationMethod(method='EquiptmentLevel',floorArea=False,numPeople=False)
        fridge.setDesignLevel(450)
        fridge.setFractionRadiant(1.0)

        # Create Fridge Instance
        space_1_fridge = openstudio.model.ElectricEquipment(fridge)
        space_1_fridge.setSpace(space)
        space_1_fridge.setSchedule(fridge_schedule)

        # Create lights definition
        lights = openstudio.model.LightsDefinition(model)
        lights.setName(str(name) +' Unit Lights')
        lights.setDesignLevelCalculationMethod(method='LightingLevel',floorArea=False,numPeople=False)
        lights.setLightingLevel(450)

# Create Schedule Set
# unit_schedule_set = openstudio.model.DefaultScheduleSet(model)
# unit_schedule_set.setName('Unit Schedules')




# Create Lights Schedule

# Lighting Compact Schedule
lights_schedule = openstudio.model.ScheduleCompact(model)
lights_schedule.setName('Unit Lighting Schedule')
lights_schedule.setString(value='Through: 12/31', index=3)
lights_schedule.setString(value='For: AllOtherDays', index=4)
lights_schedule.setString(value='Until: 24:00', index=5)
lights_schedule.setDouble(value=1.0, index=6)

def hourSch(model,nameSch, hourlyValues):
    lights_schedule = openstudio.model.ScheduleCompact(model)
    lights_schedule.setName('Unit Lighting Schedule')
    lights_schedule.setString(value='Through: 12/31', index=3)
    lights_schedule.setString(value='For: WinterDesignDay', index=4)
    lights_schedule.setString(value='Until: 24:00', index=5)
    lights_schedule.setDouble(value=0, index=6)
    lights_schedule.setString(value='For: SummerDesignDay', index=7)
    lights_schedule.setString(value='Until: 24:00', index=8)
    lights_schedule.setDouble(value=1.0, index=9)
    lights_schedule.setString(value='For: AllOtherDays', index=10)

    params = [x for x in hourlyValues]
    schValues = []
    count = 10
    for i,param in enumerate(params):
        count = count + 1
        lights_schedule.setString(value=('Until: ' + str(i + 1) + ':00'), index=count)
        count = count + 1
        lights_schedule.setDouble(value=param,index=count)
    
SchName_Lighting = 'Phius_Lighting'
SchValues_Lighting = [0.008, 0.008, 0.008, 0.008, 0.024, 0.050, 0.056, 0.050, 0.022, 0.015, 0.015, 0.015, 0.015, 0.015, 0.026, 0.015, 0.056, 0.078, 0.105, 0.126, 0.128, 0.088, 0.049, 0.020]

hourSch(model,SchName_Lighting,SchValues_Lighting)

# space_1_lights = openstudio.model.Lights(lights)
# space_1_lights.setSpace(space_1)
# space_1_lights.setSchedule(lights_schedule)

# print(model)

print(model)
```
