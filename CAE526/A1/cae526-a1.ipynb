#Problem 3


```python
data = pd.read_csv("CAE526_A1 - Problem 3.csv", index_col='Building')
e1s = list(data.loc['E1 S'])
hhs = list(data.loc['HH S'])
temp = list(data.loc['AvgTempF'])

fig, ax = plt.subplots(figsize=(6.5,4))
ax.scatter(temp, e1s, label='E1 Steam')
ax.scatter(temp, hhs, label='HH Steam')

ax.set_xlabel('Avg Monthly Temperature [F]')
ax.set_ylabel('Montly Consumption [kBtu/sf]')
ax.set_title('Montly Steam Consumption')
ax.legend()

e1e = list(data.loc['E1 Elec'])
hhe = list(data.loc['HH Elec'])
temp = list(data.loc['AvgTempF'])

fig, ax2 = plt.subplots(figsize=(6.5,4))
ax2.scatter(temp, e1e, label='E1 Electricity')
ax2.scatter(temp, hhe, label='HH Electricity')

ax2.set_xlabel('Avg Monthly Temperature [F]')
ax2.set_ylabel('Montly Consumption [kBtu/sf]')
ax2.set_title('Montly Electricity Consumption')
ax2.legend()

e1c = list(data.loc['E1 CHW'])
hhc = list(data.loc['HH CHW'])
temp = list(data.loc['AvgTempF'])

fig, ax3 = plt.subplots(figsize=(6.5,4))
ax3.scatter(temp, e1c, label='E1 CHW')
ax3.scatter(temp, hhc, label='HH CHW')

ax3.set_xlabel('Avg Monthly Temperature [F]')
ax3.set_ylabel('Montly Consumption [kBtu/sf]')
ax3.set_title('Montly Chilled Water Consumption')
ax3.legend()

data.head(10)
```
