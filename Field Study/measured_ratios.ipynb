```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_a = pd.read_excel("D:/School/IIT-PHD/FIeld Study/Data/0_A_Complete Data.xlsx")
df_t = pd.read_excel("D:/School/IIT-PHD/FIeld Study/Data/0_T_Complete Data.xlsx")
df_r = pd.read_excel("D:/School/IIT-PHD/FIeld Study/Data/0_R_Complete Data.xlsx")

heating_setpoint = 22.2

a_o_t = df_a['A_O_T [°C]']
a_1_n_t = df_a['A_1_N_T  [°C]']
a_1_s_t = df_a['A_1_S_T  [°C]']
a_2_n_t = df_a['A_2_N_T  [°C]']
a_2_s_t = df_a['A_2_S_T  [°C]']
a_3_n_t = df_a['A_3_N_T  [°C]']
a_3_s_t = df_a['A_3_S_T  [°C]']

a_24hr_in = []
a_24hr_out = []

for hour in range(0,1439):
    a_24hr_out.append(heating_setpoint - a_o_t[hour])

    indoors = []

    indoors.append(a_1_n_t[hour])
    indoors.append(a_1_s_t[hour])
    indoors.append(a_2_n_t[hour])
    indoors.append(a_2_s_t[hour])
    indoors.append(a_3_n_t[hour])
    indoors.append(a_3_s_t[hour])

    a_24hr_in.append(heating_setpoint - np.average(indoors))

print('Ratio A = ' + str(round(sum(a_24hr_in)/sum(a_24hr_out),3)))

t_o_t = df_t['T_O_T [°C]']
t_1_n_t = df_t['T_1_W_T  [°C]']
t_1_s_t = df_t['T_1_E_T  [°C]']
t_2_n_t = df_t['T_2_W_T  [°C]']
t_2_s_t = df_t['T_2_E_T  [°C]']
t_3_n_t = df_t['T_3_W_T  [°C]']
t_3_s_t = df_t['T_3_E_T  [°C]']

t_24hr_in = []
t_24hr_out = []

for hour in range(0,1439):
    t_24hr_out.append(heating_setpoint - t_o_t[hour])

    indoors = []

    indoors.append(t_1_n_t[hour])
    indoors.append(t_1_s_t[hour])
    indoors.append(t_2_n_t[hour])
    indoors.append(t_2_s_t[hour])
    indoors.append(t_3_n_t[hour])
    indoors.append(t_3_s_t[hour])

    t_24hr_in.append(heating_setpoint - np.average(indoors))

print('Ratio T = ' + str(round(sum(t_24hr_in)/sum(t_24hr_out),3)))

r_o_t = df_r['R_O_T [°C]']
r_1_n_t = df_r['R_1_W_T  [°C]']
# r_1_s_t = df_r['R_1_E_T  [°C]']
r_2_n_t = df_r['R_2_W_T  [°C]']
r_2_s_t = df_r['R_2_E_T  [°C]']
r_3_n_t = df_r['R_3_W_T  [°C]']
r_3_s_t = df_r['R_3_E_T  [°C]']

r_24hr_in = []
r_24hr_out = []

for hour in range(0,1439):
    r_24hr_out.append(heating_setpoint - r_o_t[hour])

    indoors = []

    indoors.append(r_1_n_t[hour])
    # indoors.append(r_1_s_t[hour])
    indoors.append(r_2_n_t[hour])
    indoors.append(r_2_s_t[hour])
    indoors.append(r_3_n_t[hour])
    indoors.append(r_3_s_t[hour])


    r_24hr_in.append(heating_setpoint - np.average(indoors))

print('Ratio R = ' + str(round(sum(r_24hr_in)/sum(r_24hr_out),3)))



```
