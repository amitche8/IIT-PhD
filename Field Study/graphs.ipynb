```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("D:/School/IIT-PHD/FIeld Study/Data/0_A_Complete Data.xlsx")

t = df['Date-Time (CDT)']
a_o_t = df['A_O_T [C]']
a_1_n_t = df['A_1_N_T  [C]']
a_1_s_t = df['A_1_S_T  [C]']
a_2_n_t = df['A_2_N_T  [C]']
a_2_s_t = df['A_2_S_T  [C]']
a_3_n_t = df['A_3_N_T  [C]']
a_3_s_t = df['A_3_S_T  [C]']


a_1_n_mrt = df['A_1_N_MRT  [C]']
a_1_s_mrt = df['A_1_S_MRT  [C]']
a_2_n_mrt = df['A_2_N_MRT  [C]']
a_2_s_mrt = df['A_2_S_MRT  [C]']
a_3_n_mrt = df['A_3_N_MRT  [C]']
a_3_s_mrt = df['A_3_S_MRT  [C]']

a_o_rh = df['A_O_RH [%]']
a_1_n_rh = df['A_1_N_RH  [%]']
a_1_s_rh = df['A_1_S_RH  [%]']
a_2_n_rh = df['A_2_N_RH  [%]']
a_2_s_rh = df['A_2_S_RH  [%]']
a_3_n_rh = df['A_3_N_RH  [%]']
a_3_s_rh = df['A_3_S_RH  [%]']

fig, axs = plt.subplots(3, 1, layout='constrained', figsize=(8.5,11))
axs[0].plot(t, a_o_t, linewidth=2, linestyle='dashed', label='Outdoor')
axs[0].plot(t, a_1_n_t, linewidth=0.6, color = 'darkgreen', label='Basement N')
axs[0].plot(t, a_1_s_t, linewidth=0.6, color= 'chartreuse', label='Basement S')
axs[0].plot(t, a_2_n_t, linewidth=0.6, color= 'mediumblue', label='1F N')
axs[0].plot(t, a_2_s_t, linewidth=0.6, color= 'cornflowerblue', label='1F S')
axs[0].plot(t, a_3_n_t, linewidth=0.6, color= 'darkviolet', label='2F N')
axs[0].plot(t, a_3_s_t, linewidth=0.6, color= 'orchid', label='2F S')

axs[1].plot(t, a_1_n_mrt, linewidth=0.6, color = 'darkgreen', label='Basement N')
axs[1].plot(t, a_1_s_mrt, linewidth=0.6, color= 'chartreuse', label='Basement S')
axs[1].plot(t, a_2_n_mrt, linewidth=0.6, color= 'mediumblue', label='1F N')
axs[1].plot(t, a_2_s_mrt, linewidth=0.6, color= 'cornflowerblue', label='1F S')
axs[1].plot(t, a_3_n_mrt, linewidth=0.6, color= 'darkviolet', label='2F N')
axs[1].plot(t, a_3_s_mrt, linewidth=0.6, color= 'orchid', label='2F S')

axs[2].plot(t, a_o_rh, linewidth=2, linestyle='dashed', label='Outdoor')
axs[2].plot(t, a_1_n_rh, linewidth=0.6, color = 'darkgreen', label='Basement N')
axs[2].plot(t, a_1_s_rh, linewidth=0.6, color= 'chartreuse', label='Basement S')
axs[2].plot(t, a_2_n_rh, linewidth=0.6, color= 'mediumblue', label='1F N')
axs[2].plot(t, a_2_s_rh, linewidth=0.6, color= 'cornflowerblue', label='1F S')
axs[2].plot(t, a_3_n_rh, linewidth=0.6, color= 'darkviolet', label='2F N')
axs[2].plot(t, a_3_s_rh, linewidth=0.6, color= 'orchid', label='2F S')

# axs[0].set_xlim(0, 2)
axs[0].set_xlabel('Date')
axs[1].set_xlabel('Date')
axs[2].set_xlabel('Date')
axs[0].set_ylabel('Temperature [C]')
axs[1].set_ylabel('RadiantTemperature [C]')
axs[2].set_ylabel('Relative Humidity [%]')
axs[0].grid(True)
axs[1].grid(True)
axs[2].grid(True)
axs[0].legend(loc='lower left', fontsize=6)
axs[1].legend(loc='lower left', fontsize=6)
axs[2].legend(loc='lower left', fontsize=6)


plt.suptitle('House A')
plt.show()
```
