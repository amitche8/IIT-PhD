```python
import matplotlib.pyplot as plt
import numpy as np

species = ("Thermal Comfort Normal", "Clothing Normal", "Thermal Comfort Outage", "Clothing Outage")
penguin_means = {
    'Low Comfort / Clothing': (0,0,0,0),
    'Medium Comfort / Clothing': (0,2,0,0),
    'Medium - High Comfort / Clothing': (1,4,6,6),
    'High Comfort / Clothing': (5,0,0,0),
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Respondents')
ax.set_xlabel('Survey Topic')
ax.set_title('Occupant Survey Data')
ax.set_xticks(x + width, species, rotation=20)
ax.legend(loc='best', fontsize=7)
ax.set_ylim(0, 7)

plt.show()
```
